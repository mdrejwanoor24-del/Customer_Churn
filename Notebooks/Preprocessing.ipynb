# Data Pre-processing
-Data type correction
-Duplicate handling 
-Remove irregular columns
-Saved clean data


```python
#imports
import numpy as np
import pandas as pd
```

```python
#load data\data\Telco_Customer_Churn.csv
DATA_PATH = r"C:\Users\DELL\Downloads\Customer_Churn_Prediction\data\Telco_Customer_Churn.csv"
data = pd.read_csv(DATA_PATH)
data.head()
```

```python
#data back_up
df = data.copy()
df.info()
```

```python
#meta data
print("Number of Rows:",df.shape[0])
print("Number of Columns:",df.shape[1])
all_columns = df.columns.to_list()
num_columns = df.select_dtypes(include="number").columns.to_list()
cat_columns = df.select_dtypes(include="object").columns.to_list()
print("\n All columns: \n",all_columns)
print("\n Nmerical columns:\n",num_columns)
print("\n Categorical columns:\n",cat_columns)
print("\n Data type:\n",df.dtypes)
    
```

```python
#duplicate
print("Duplicate Details:", df.duplicated().sum())
print("\n Duplicate Rows:\n")
df[df.duplicated()]
```

```python
#Drop irrelevent columns
df.drop(columns =["customerID"],axis=1,inplace= True)
```

```python
#change data type of TotalCharges
df["TotalCharges"]= pd.to_numeric(df["TotalCharges"],errors="coerce")
df["TotalCharges"].dtype
```

```python
#Saved cleaned data
CEANED_DATA_PATH = r"C:\Users\DELL\Downloads\Customer_Churn_Prediction\Data\cleaned_data.csv"
df.to_csv(CEANED_DATA_PATH,index= False)
```
